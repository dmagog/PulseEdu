````python
with open('./history_tests.json', 'r') as f:
    test_tasks = json.load(f)

with open('./history_student_test_answers.json', 'r') as f:
    student_test_answers = json.load(f)
````

````python
# import pandas as pd
answers_by_topic = {}

topics = []

for test in test_tasks:
    topic = test['Тема']
    topics.append(topic)
    answers = []
    for task in test['Задания']:
        answers.append(task['Правильный ответ'])
    answers_by_topic[topic] = answers

student_answers_by_topic = {}

for student_data in student_test_answers:
    student_id = student_data['Студент']
    student_answers = {}
    for test_answers in student_data['Ответы']:
        topic = test_answers['Тема']
        answers = test_answers['Ответы']
        student_answers[topic] = answers
    student_answers_by_topic[student_id] = student_answers
````

````python
from collections import defaultdict

df_data = pd.DataFrame({
    'topic': topics,
    'correct_answers': [answers_by_topic[topic] for topic in topics]
})

for student_id, answers in student_answers_by_topic.items():
    df_data[f'student_{student_id}'] = [answers[topic] for topic in topics]

# Давайте посчитаем только задания с ответом, в которых требуется выбор правильного ответа из вариантов, и посмотрим качество на них
df_data['task_cnt'] = df_data['correct_answers'].apply(lambda x: sum(len(y) == 1 for y in x))

student_accuracy_list = defaultdict(list)
for idx in range(1, 11):
    student_id = f'student_{idx}'
    for correct_answers, student_answers in zip(df_data['correct_answers'].to_numpy(), df_data[student_id].to_numpy()):
        task_cnt = 0
        correct_ans_cnt = 0
        for correct_answer, student_answer in zip(correct_answers, student_answers):
            if len(correct_answer) == 1:
                task_cnt += 1
                correct_ans_cnt += correct_answer == student_answer
        student_accuracy_list[student_id].append(round(correct_ans_cnt / task_cnt, 2))

print(df_data['task_cnt'].to_numpy())
for student_id, accuracy_list in student_accuracy_list.items():
    print(f'{student_id} - {accuracy_list}')
````

````python
TEST_FEEDBACK_PROMPT = """
# Система обратной связи студентам

Ты - ассистент в системе обратной связи для студентов. Ты эксперт в той теме, которую проверяешь. Твоя задача: проанализировать задания, выданные студенту университета, и его ответы на данный тест. По итогам проверки тебе необходимо дать обратную связь для студента.

## Критерии оценки

Задания теста можно разбить на 2 типа: в одном студент выбирает правильный ответ из множества предоставленных ему вариантов, другие - задания с развернутым ответом. Внимательно читай развернутые ответы и принимай решение об их правильности. Если в качестве ответа дана пустая строка, то это означает, что студент пропустил вопрос. Постарайся логически связать пропуски и ошибки студента и дать полезный комментарий, который поможет восполнить студенту пробелы в знаниях.

С правильными ответами в заданиях спорить нельзя!

В заданиях с развернутым ответом не обязательно дословное совпадение. Проанализируй ответа студента и оцени, схож ли он по смыслу с правильным ответом. Если да, то такой ответ можно засчитывать как правильный.

## Рекомендации к обратной связи

Дай комментарий как эксперт в теме теста, укажи студенту на его ошибки и подскажи, что ему нужно подтянуть. Оценивай результат студента объективно. В заданиях с развернутым ответом кратко объясни почему студент не прав и почему его ответ не засчитан.
Постарайся обобщить проблему ответов студента в обратной связи и дать более полный комментарий.

Выдай ответ в формате JSON:
```json
{
"mistakes": "Список номеров заданий, в которых студентом были допущены ошибки. Внимательно проверять, что выделил нужные номера заданий.",
"advice": "Твоя рекомендация для студента. Считай, что ты обращаешься лично к нему. Для ошибок в заданиях с развернутым ответом постарайся также объяснить студенту, почему он не прав."
}
```
Пример:
{
"mistakes": ["1", "3", "5"],
"advice": "Вам необходимо повторить тему неопределенных интегралов."
}

## Структура входных данных

Тебе будут даны 2 таблицы в формате JSON. Первая таблица описывает задания теста с ответами, вторая содержит ответы студента.

Формат 1 таблицы:
Таблица содержит список заданий. Каждое задание описывается в формате JSON следующим образом:
```json
"Вопрос": "Вопрос, который задан студенту",
"Варианты ответов": "Список вариантов ответов, предложенных студенту. Если этого поля нет, то это задание с развернутым ответом.",
"Правильный ответ": "Правильный ответ на тест. Если есть варианты ответов, то это порядковый номер правильного варианта, иначе - развернутый ответ на вопрос."
```

Формат 2 таблицы:
Таблица содержит список ответов студента. Ответ на каждое задание представлен в виде строки. Если строка пустая, значит студент не ответил на данный вопрос.
""".strip()

def get_test_feedback(model, test_tasks, student_answers):
    test_tasks_json = json.dumps(test_tasks, indent=0, ensure_ascii=False)
    student_answers_json = json.dumps(student_answers, indent=0, ensure_ascii=False)

    user_data = "\n".join(["Входные данные:", "1 таблица", test_tasks_json, "2 таблица", student_answers_json])

    messages = [
        {"role": "system", "text": TEST_FEEDBACK_PROMPT},
        {"role": "user", "text": user_data},
    ]

    return model.run(messages=messages)
````
